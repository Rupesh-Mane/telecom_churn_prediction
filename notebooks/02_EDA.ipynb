IMPORT LIBRARIES

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

LOAD DATASET


```python
df=pd.read_csv("../data/raw/telecom_customer_churn.csv")
```

COPY THE DATASET

```python
df_eda=df.copy()
```

CHECK CHURN COUNT

```python
df_eda["Churn"].value_counts()
```

PLOT CHURN DISTRIBUTION

```python
sns.countplot(data=df_eda,x="Churn")
plt.title("Customer churn Distribution")
plt.show()
```

Analyze--> Tenure

```python
"""
We want to understand:

How long have customers been with the telecom company?
"""
```

SEE THE DISTRIBUTION OF TENURE

```python
sns.histplot(data=df_eda,x="Tenure")
plt.title("Tenure Distribution")
plt.show()
```

COMPARE TENURE WITH CHURN

```python
sns.boxplot(data=df_eda, x="Churn", y="Tenure")
plt.title("Tenure vs Churn")
plt.show()
```
